```python
# this is for "read the txt.file" 
# I split the independent indexs and data
## you can check the total_list and indexs before you do this project.
## this is the given instructions, I check where you have to write some codes.
import random 

f = open("Bank_account.txt", 'r', encoding='utf8')
total_list = []
indexs = f.readline()[:-1]
temps = indexs
print(temps)
while(temps != ''):
    temps = f.readline()[:-1]
    total_list.append(temps)
f.close()    
  
print("1. Create new account.")
print("2. For transactions.")
print("3. Check the details of existing account.")
print("4. View customer's list.")
print("5. Exit.")


## 여기에서는 choose the option(1~5) 을 input으로 받는 것을 작성해 보세요!

while(True):
    number = input("Choose the option (1~5): ")
    if number in ['1','2','3','4','5']:
        break





## 여기에서는 input이 1번일때의, 코드를 작성해보세요!
###cf) if number == '1':
if number =='1':
    name = input("Name: ")
    birth = input("Birth date (yymmdd): ")
    passwd = input("Passwd(4 digis):")
    
    account_list = []
    random_account = 0
    for i in range(len(total_list)-1):
        account_list.append(total_list[i].split()[3])
        
    while(True):
        random_account = random.randint(100000,999999)
        if random_account not in  account_list:
            break
        
        
    
    balance = input("The amount you want to deposit first :")
    
    wrapping = "{} {} {} {} {}".format(name,birth,passwd,random_account,balance)
    total_list[-1] = wrapping
    total_list.append("")

## 여기에서는 input이 2번일때의, 코드를 작성해보세요!
###cf) if number == '2':
if number == '2':
    name = input("What's your name? ")
    while(True):
        option = input("1. Deposit  2. Withdraw: ")
        if option in ['1','2']:
            break
    
    finding_index = -1
    for i in range(len(total_list)-1):
        if total_list[i].split()[0] == name:
            finding_index = i
            
    if finding_index == -1:
        print("There is no account in our bank")
        
    else:
        line = total_list[finding_index].split()

        if option =='1' :
            deposit = input("How much would you like to deposit?: ")
            new_account = int(deposit)+int(line[4])
            print("Your balance after transaction is {}".format(new_account))



            wrapping = "{} {} {} {} {}".format(line[0],line[1],line[2],line[3],str(new_account))
            total_list[finding_index] = wrapping

        elif option == '2':
            exist_money = int(line[4])
            withdraw = 0
            while(True):
                print("Your balance is {}".format(exist_money))
                withdraw = int(input("How much would you like to withdraw: "))

                if exist_money < int(withdraw):
                    print("Insufficient Balance...")

                else:
                    break
            new_account = exist_money - withdraw
            print("Your balance after transaction is {}".format(new_account))
            wrapping = "{} {} {} {} {}".format(line[0],line[1],line[2],line[3],str(new_account))
            total_list[finding_index] = wrapping
        

## 여기에서는 input이 3번일때의, 코드를 작성해보세요!
###cf) if number == '3':
if number == '3':
    input_3 = input("What is your name? ")
    print("Your name is {}".format(input_3))
    
    finding_index = -1
    for i in range(len(total_list)-1):
        if total_list[i].split()[0] == input_3:
            finding_index = i
            
    if finding_index == -1:
        print("There is no account in our bank")
    
    else:
        reset_num = 1
        while(reset_num<6):
            password = input("Enter your password(4 digits): ")
            if password == total_list[finding_index].split()[2]:
                print("Your account number is {}".format(total_list[finding_index].split()[3]))
                print("Your balance is {}".format(total_list[finding_index].split()[4]))
                break
            reset_num +=1
                
        if (reset_num ==6):
            print("You have entered wrong password 5 times, shut down the system...")


## 여기에서는 input이 4번일때의, 코드를 작성해보세요!
###cf) if number == '4':
if number == '4':
    for i in range(len(total_list)-1):
        print(total_list[i].split()[0])



# 여기에서는 input이 5번일때의, 코드를 작성해보세요!
## input이 5번일때에는, Exit 이기 때문에, 아무 작업을 안해도 됩니다.
###cf) if number == '5':
    

    
# this is for "write the txt.file" 
## 1~5 transaction 을 통해, 수정된 내용을 반영할 수도 있고 (1~2)
## 기존의 내용을 그대로 다시 작성할 수도 있습니다. (3~5)

# format 을 유지하는것이 중요합니다.
## 기존에, given format이, (indexs 와 total_list가 어떤 방식으로 적혀있는지를 파악하시길 바랍니다.)

f = open("Bank_account.txt", 'w', encoding='utf8')
f.write(indexs+ '\n')
for i in total_list:
    f.write(i + '\n')
f.close()
```
